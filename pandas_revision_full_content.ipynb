```python
# merging
students_df = pd.DataFrame({
    'student_id': [1, 2, 3, 4,5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', "Tom"],
    'age': [20, 21, 22, 23, 25]
})
grades_df = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'course': ['Math', 'History', 'Science', 'Art'],
    'grade': ['A', 'B+', 'A-', 'B']
})
```

```python
df = pd.DataFrame({
    'date': ['2023-01', '2023-02'],
    'Kigali_temp': [25, 27],
    'Kigali_humid': [60, 55],
    'Musanze_temp': [20, 22],
    'Musanze_humid': [65, 62]
})
```

```python
temp = df.melt(id_vars='date', value_vars=['Kigali_temp', 'Musanze_temp'],
               var_name='city', value_name='temperature')
temp['city'] = temp['city'].str.replace('_temp', '')
temp
```

```python
humid= df.melt (id_vars="date", value_vars= ["Kigali_humid", "Musanze_humid"], var_name= "city", value_name= "humidity")
humid["city"]= humid["city"].str.replace("_humid", "")
humid
```

```python
new_formed= pd.merge(temp, humid, on=["date", "city"])
new_formed
```
